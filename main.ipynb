```python
import pandas as pd

df = pd.read_csv(r"D:\s\updatedstationshour.csv")
df
```

```python
df.info()
```

```python
# AQI_Bucket_Dict = df['AQI_Bucket'].value_counts()
City_Dict = df['City'].value_counts()
Date_Dict = df['Date'].value_counts()
State_Dict = df['State'].value_counts()
StationId_Dict = df['StationId'].value_counts()
Time_Dict = df['Time'].value_counts()




# df['AQI_Bucket_Dict'] = df['AQI_Bucket'].map(AQI_Bucket_Dict)
df['City_Dict'] = df['City'].map(City_Dict)
df['Date_Dict'] = df['Date'].map(Date_Dict)
df['State_Dict'] = df['State'].map(State_Dict)
df['StationId_Dict'] = df['StationId'].map(StationId_Dict)
df['Time_Dict'] = df['Time'].map(Time_Dict)
```

```python
print(f"AQI_Bucket_Dict:{AQI_Bucket_Dict}, City_Dict: {City_Dict},Date_Dict:{Date_Dict},State_Dict:{State_Dict},StationId_Dict:{StationId_Dict},Time_Dict:{Time_Dict}")
```

```python
df['AQI_Bucket'].unique()
```

```python
from sklearn import preprocessing
le = preprocessing.LabelEncoder()
df['AQI_Bucket'] = le.fit_transform(df['AQI_Bucket'])
# df['City'] = le.fit_transform(df['City'])
# df['State'] = le.fit_transform(df['State'])
# df['StationId'] = le.fit_transform(df['StationId'])
# df['Date'] = le.fit_transform(df['Date'])
# df['Time'] = le.fit_transform(df['Time'])
```

```python
n_classes = len(df['AQI_Bucket'].unique())
n_classes
```

```python
df['AQI_Bucket'].unique()
```

```python
col_names = ['City','Date','State','StationId','Time','AQI']
df = df.drop(col_names,axis=1)
df
```

```python
X = df.drop('AQI_Bucket',axis=1)
```

```python
y = df['AQI_Bucket']
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test= train_test_split(X,y, test_size = 0.2, shuffle= True)
```

```python
from xgboost import XGBClassifier
```

```python
bst = XGBClassifier(n_estimators=100, max_depth=6, learning_rate=0.0001, objective='multi:softmax',num_class=n_classes,eval_metric= 'rmse')
# fit model
bst.fit(X_train, y_train)
```
