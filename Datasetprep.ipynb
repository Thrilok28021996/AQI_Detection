```python
import pandas as pd
```

```python
city_daydata = pd.read_csv(r"D:\s\AQI(2015-2020)\city_day.csv")
city_daydata = city_daydata.dropna()
city_daydata.head()
```

```python
city_daydata.isnull().sum()
```

```python
city_hourdata = pd.read_csv(r"D:\s\AQI(2015-2020)\city_hour.csv")
city_hourdata = city_hourdata.dropna()
city_hourdata.Datetime = city_hourdata.Datetime.apply(pd.to_datetime)
city_hourdata['Date'] = [d.date() for d in city_hourdata['Datetime']]
city_hourdata['Time'] = [d.time() for d in city_hourdata['Datetime']]
city_hourdata = city_hourdata.drop('Datetime', axis=1)
city_hourdata.head()
```

```python
city_hourdata['City'].unique()
```

```python
city_hourdata.isnull().sum()
```

```python
station_daydata = pd.read_csv(r"D:\s\AQI(2015-2020)\station_day.csv")
station_daydata = station_daydata.dropna()
station_daydata.head()
```

```python
station_daydata.isnull().sum()
```

```python
station_hourdata = pd.read_csv(r"D:\s\AQI(2015-2020)\station_hour.csv")
station_hourdata = station_hourdata.dropna()
station_hourdata.Datetime = station_hourdata.Datetime.apply(pd.to_datetime)
station_hourdata['Date'] = [d.date() for d in station_hourdata['Datetime']]
station_hourdata['Time'] = [d.time() for d in station_hourdata['Datetime']]
station_hourdata = station_hourdata.drop('Datetime', axis=1)
station_hourdata.head()
```

```python
station_hourdata.isnull().sum()
```

```python
stations_data = pd.read_csv(r"D:\s\AQI(2015-2020)\stations.csv")
stations_data = stations_data.drop('Status', axis=1)
stations_data = stations_data.drop('StationName', axis=1)
stations_data.head()
```

```python
stations_data.isnull().sum()
```

```python
print(f'The length of stations data: {len(stations_data)}')
print(f'The length of station hour data: {len(station_hourdata)}')
print(f'The length of station day data: {len(station_daydata)}')
print(f'The length of city hour data: {len(city_hourdata)}')
print(f'The length of city day data: {len(city_daydata)}')
```

```python
updatedstationshour_data = pd.merge(stations_data, station_hourdata, on='StationId', how='outer')
updatedstationshour_data = updatedstationshour_data.dropna()
updatedstationshour_data = updatedstationshour_data[sorted(updatedstationshour_data)]
updatedstationshour_data.to_csv("updatedstationshour.csv",index = False)
updatedstationshour_data.head()
```

```python
print(f'The length of updated station hour data: {len(updatedstationshour_data)}')
```

```python
updatedstationsday_data = pd.merge(station_daydata, stations_data, on='StationId', how='outer')
updatedstationsday_data = updatedstationsday_data.dropna()
updatedstationsday_data = updatedstationsday_data[sorted(updatedstationsday_data)]

updatedstationsday_data.head()
```

```python
print(f'The length of updated station hour data: {len(updatedstationsday_data)}')
```
